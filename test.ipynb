```python
# Imported in the fishplot.py module does not work in this notebook.
from fishplot import fishplot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# the dataframe here comes from direct import of a dictionary
# a easier way to collect data is using excel file and then save as .csv
# 3 columns: col-1: date; col-2: variant/alteration; col-3: percentage of each variant/alteration at each time point
# for the date column, every mutation/alteration needs a time point 
# (e.g., if we have 3 mutations at one testing date, we need 3 rows with the same date)

d = {'date':['2010/09/20','2010/09/20','2010/10/20','2010/10/20','2011/11/20','2011/11/20'],
            'variants':['TP53','NTRK','TP53','NTRK','TP53','NTRK'],
            'freq':[0.3,0.2,0.1,0.2,0.2,0.5]}
df = pd.DataFrame(data=d)

print(df)
df.to_csv('test')
fishplot(df)
plt.legend(labels=['TP53', 'NTRK'], loc='upper left')
plt.show()
```
